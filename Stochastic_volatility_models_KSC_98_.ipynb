### Build Model

```python
import pytensor
from pymc.pytensorf import collect_default_updates

T = len(df)

def h_dist(h_init, mu, phi, sigma, size):
    def one_step(h_tm1, mu, phi, sigma):
        mu_ = mu + phi * (h_tm1 - mu)
        h_t = pm.Normal.dist(mu_, sigma)
        return h_t, collect_default_updates(outputs=[h_t])

    h, updates = pytensor.scan(fn=one_step,
                                outputs_info=[h_init],
                                non_sequences=[mu, phi, sigma],
                                n_steps=T,
                                name='h',
                                strict=True)

    return h

def make_stochastic_volatility_model(data):
    coords={'Datetime': data['Datetime'].values}
    with pm.Model(coords=coords) as model:

        mu = pm.Cauchy('mu', alpha=0., beta=10.)
        phi = pm.Uniform('phi', lower=-1., upper=1.)
        sigma = pm.HalfCauchy('sigma', beta=5.)

        h_init = pm.Normal('h_init', mu=mu, sigma=(sigma / pm.math.sqrt(1 - phi**2)))

        h_innov = pm.CustomDist('h_innov',
                                h_init,
                                mu,
                                phi,
                                sigma,
                                dist=h_dist,
                                dims='Datetime')

        h = pm.Deterministic(name='h', var=pt.concatenate([h_init[None], h_innov[:-1]], axis=0), dims='Datetime')

        # h = pm.Deterministic('h_dist', h_dist(h_init, mu, phi, sigma, size=n_steps))
        y_obs = pm.Normal('y_obs', mu=0., sigma=pm.math.exp(h / 2), observed=df['Diff'], dims='Datetime')

    return model

stochastic_vol_model = make_stochastic_volatility_model(df)
```

### Check the model

```python
rng = np.random.RandomState(1234)
with stochastic_vol_model:
    trace = pm.sample_prior_predictive(samples=500, random_seed=rng)

prior_predictive = az.extract(trace, group='prior_predictive')
```

```python
rng = np.random.RandomState(1234)
with stochastic_vol_model:
    trace.extend(pm.sampling.jax.sample_numpyro_nuts(random_seed=rng, chain_method='parallel'))

posterior = az.extract(trace)
posterior['exp_h'] = np.exp(posterior['h'])
```
